```python
import pandas as pd
import numpy as np
```

```python
x = np.array([3,4.3,5,7.5,9]).reshape(5,1)
y = np.array([4.3,5,6.6,7,8.7]).reshape(5,1)
```

```python
x
```

```python
y
```

```python
a=1.5
b=1.7
lr=0.001
for i in range(50):
  yh=a*x+b
  dea= -2*((y-yh)*x).sum()
  deb= -2*(y-yh).sum()
  a=a-lr*dea

  b=b-lr*deb
  
  error=((y-yh)**2).mean() 
  print(error)
print("a:",a)  
print("b:",b)  
```
