```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d=pd.read_csv("googleplaystore.csv", nrows=1000)
df=d.copy()
df
```

```python
df.shape
```

```python
x=df["Rating"]
y=df["Reviews"]
plt.scatter(x, y)
```

```python
plt.scatter(x, y, color="r")

plt.title("Google PlayStore Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.scatter(x, y, color="r", marker="*")

plt.title("Google PlayStore Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.scatter(x, y, color="r", marker="*", s=100)

plt.title("Google PlayStore Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.scatter(x, y, color="r", marker="*", s=100, alpha=0.5)

plt.title("Google PlayStore Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.scatter(x, y, color="r", marker="*", s=100, alpha=0.5, edgecolors="black")

plt.title("Google PlayStore Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.scatter(x, y, color="r", marker="*", s=100, alpha=0.5, edgecolors="black")

plt.scatter(x, df["Installs"], color="y", marker="o", s=100, alpha=0.5, edgecolors="red")


plt.title("Google PlayStore Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
```
