```python
import matplotlib.pyplot as plt
```

```python
clas=["Python", "R", "ML", "AI", "DS"]
student=[45, 15, 35, 25, 30]
```

```python
plt.pie([1])
```

```python
plt.pie(student)
```

```python
plt.pie(student)
plt.show()
```

```python
plt.pie(student, labels=clas)
plt.show()
```

```python
explode=[0, 0, 0, 0, 0]

plt.pie(student, labels=clas, explode=explode)
plt.show()
```

```python
explode=[0.2, 0, 0, 0, 0]

plt.pie(student, labels=clas, explode=explode)
plt.show()
```

```python
explode=[0.2, 0, 0.1, 0, 0]

plt.pie(student, labels=clas, explode=explode)
plt.show()
```

```python
explode=[0.2, 0, 0.1, 0, 0]
colors=["red", "green", "yellow", "blue", "c"]

plt.pie(student, labels=clas, explode=explode, colors=colors)
plt.show()
```

```python
explode=[0.2, 0, 0.1, 0, 0]
colors=["red", "green", "yellow", "blue", "c"]

plt.pie(student, labels=clas, explode=explode, colors=colors, autopct="%0.1f%%")
plt.show()
```

```python
explode=[0.2, 0, 0.1, 0, 0]
colors=["red", "green", "yellow", "blue", "c"]

plt.pie(student, labels=clas, explode=explode, colors=colors, autopct="%0.1f%%",
        shadow=True)
plt.show()
```

```python
explode=[0.2, 0, 0.1, 0, 0]
colors=["red", "green", "yellow", "blue", "c"]

plt.pie(student, labels=clas, explode=explode, colors=colors, autopct="%0.1f%%",
        shadow=True, startangle=90)
plt.show()
```

```python
explode=[0.2, 0, 0.1, 0, 0]
colors=["red", "green", "yellow", "blue", "c"]
textprops={"fontsize":15}

plt.pie(student, labels=clas, explode=explode, colors=colors, autopct="%0.1f%%",
        shadow=True, startangle=90, textprops=textprops)
plt.show()
```
