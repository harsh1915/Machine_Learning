```python
from matplotlib import style
style.use("ggplot")
plt.plot(days, temperature, "yo--", linewidth=4, markersize=10)
plt.title("Amdavad Temperature", fontsize=20)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.legend(["Tem Line"], loc=2)
plt.grid(color="r")
```

```python
from matplotlib import style
style.use("ggplot")
plt.plot(days, temperature, "yo:", linewidth=4, markersize=10)
plt.plot(days, mumbai_temperature, "ro--", linewidth=4, markersize=10)
plt.legend(["Amdavad", "Mumbai"])
plt.title("Amdavad & Mumbai Temperature", fontsize=20)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.grid(color="r")
```
