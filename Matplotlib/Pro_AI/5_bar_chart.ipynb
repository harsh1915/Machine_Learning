```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
clas=["Python", "R", "AI", "ML", "DS"]
clas1_std=[10, 12, 8, 11, 9]
clas2_std=[20, 15, 10, 25, 30]
clas3_std=[35, 10, 30, 15, 25]
```

```python
plt.bar(clas, clas1_std)
```

```python
plt.barh(clas, clas1_std)
```

```python
plt.bar(clas, clas1_std, width=0.6)
```

```python
plt.bar(clas, clas1_std, width=0.6, align="edge")
```

```python
plt.bar(clas, clas1_std, width=0.7, align="edge", color="y")
```

```python
plt.bar(clas, clas1_std, width=0.7, align="edge", color="y", edgecolor="r")
```

```python
plt.bar(clas, clas1_std, width=0.7, align="edge", color="y", edgecolor="r", linewidth=3)
```

```python
plt.bar(clas, clas1_std, width=0.7, align="edge", color="y", edgecolor="r", linewidth=3, linestyle="--")
```

```python
plt.bar(clas, clas1_std, width=0.7, align="edge", color="y", edgecolor="r", linewidth=3, linestyle="--", label="Class 1 Students")

plt.xlabel("Subject", fontsize=20)
plt.ylabel("Number Of Students", fontsize=20)
plt.title("Class 1 Students", fontsize=25)
plt.legend()
style.use("ggplot")
plt.grid(color="red")
```

```python
plt.bar(clas, clas1_std, width=0.2, align="edge", color="y", edgecolor="r", linewidth=2, linestyle="--", label="Class 1 Students")
plt.bar(clas, clas2_std, width=0.2, align="edge", color="b", edgecolor="y", linewidth=2, linestyle=":", label="Class 2 Students")


plt.xlabel("Subject", fontsize=20)
plt.ylabel("Number Of Students", fontsize=20)
plt.legend()
style.use("ggplot")
plt.grid(color="red")
```

```python
clas_index=np.arange(len(clas))

width=0.2
plt.bar(clas_index, clas1_std, width, align="edge", color="y", edgecolor="r", linewidth=2, linestyle="--", label="Class 1 Students")
plt.bar(clas_index + width, clas2_std, width, align="edge", color="b", edgecolor="y", linewidth=2, linestyle=":", label="Class 2 Students")


plt.xlabel("Subject", fontsize=20)
plt.ylabel("Number Of Students", fontsize=20)
plt.legend()
style.use("ggplot")
plt.grid(color="red")
```

```python
clas_index=np.arange(len(clas))

width=0.2
plt.bar(clas_index, clas1_std, width, align="edge", color="y", edgecolor="r", linewidth=2, linestyle="--", label="Class 1 Students")
plt.bar(clas_index + width, clas2_std, width, align="edge", color="b", edgecolor="y", linewidth=2, linestyle=":", label="Class 2 Students")

plt.xticks(clas_index, clas)

plt.xlabel("Subject", fontsize=20)
plt.ylabel("Number Of Students", fontsize=20)
plt.legend()
style.use("ggplot")
plt.grid(color="red")
```

```python
clas_index=np.arange(len(clas))

width=0.3
plt.bar(clas_index, clas1_std, width, align="edge", color="y", edgecolor="r", linewidth=2, linestyle="--", label="Class 1 Students")
plt.bar(clas_index + width, clas2_std, width, align="edge", color="b", edgecolor="y", linewidth=2, linestyle=":", label="Class 2 Students")

plt.xticks(clas_index, clas, rotation=20)

plt.xlabel("Subject", fontsize=20)
plt.ylabel("Number Of Students", fontsize=20)
plt.legend()
style.use("ggplot")
plt.grid(color="red")
```

```python
clas_index=np.arange(len(clas))

width=0.3
plt.bar(clas_index, clas1_std, width, align="edge", color="y", edgecolor="r", linewidth=2,
        linestyle="--", label="Class 1 Students")

plt.bar(clas_index + width, clas2_std, width, align="edge", color="b", edgecolor="y",
        linewidth=2, linestyle=":", label="Class 2 Students")

plt.bar(clas_index + width + width, clas3_std, width, align="edge", color="r", edgecolor="b",
        linewidth=2, label="Class 3 Students")

plt.xticks(clas_index, clas, rotation=20)

plt.xlabel("Subject", fontsize=20)
plt.ylabel("Number Of Students", fontsize=20)
plt.legend()
style.use("ggplot")
plt.grid(color="red")
```

```python
clas_index=np.arange(len(clas))

width=0.3
plt.barh(clas_index, clas1_std, width, align="edge", color="y", edgecolor="r", linewidth=2,
        linestyle="--", label="Class 1 Students")

plt.barh(clas_index + width, clas2_std, width, align="edge", color="b", edgecolor="y",
        linewidth=2, linestyle=":", label="Class 2 Students")

plt.barh(clas_index + width + width, clas3_std, width, align="edge", color="r", edgecolor="b",
        linewidth=2, label="Class 3 Students")

plt.yticks(clas_index, clas, rotation=20)

plt.ylabel("Subject", fontsize=20)
plt.xlabel("Number Of Students", fontsize=20)
plt.legend()
style.use("ggplot")
plt.grid(color="red")
```
