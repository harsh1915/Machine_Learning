```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_std_age=np.random.randint(20, 30, (100))
py_std_age=np.random.randint(20, 30, (100))
```

```python
ml_std_age
```

```python
py_std_age
```

```python
plt.hist(ml_std_age)
```

```python
plt.hist(ml_std_age)

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
```

```python
bins=[20, 22, 24, 26, 28, 30]

plt.hist(ml_std_age, bins)

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
```

```python
bins=[20, 22, 24, 26, 28, 30]

plt.hist(ml_std_age, bins, rwidth=0.8)

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
```

```python
bins=[20, 22, 24, 26, 28, 30]

plt.hist(ml_std_age, bins, rwidth=0.8, histtype="step")

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
```

```python
bins=[20, 22, 24, 26, 28, 30]

plt.hist(ml_std_age, bins, rwidth=0.8, color="r")

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
```

```python
bins=[20, 22, 24, 26, 28, 30]

plt.hist(ml_std_age, bins, rwidth=0.8, color="r", label="ML Std")

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
```

```python
bins=[20, 22, 24, 26, 28, 30]

plt.hist(ml_std_age, bins, rwidth=0.8, color="r", label="ML Std")

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
```

```python
bins=[20, 22, 24, 26, 28, 30]

plt.hist(ml_std_age, bins, rwidth=0.8, color="r", label="ML Std")
plt.hist(py_std_age, bins, rwidth=0.8, color="y", label="PY Std")

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
```

```python
bins=[20, 22, 24, 26, 28, 30]

plt.hist([ml_std_age,py_std_age], bins, rwidth=0.8, color=["r", "y"], label=["ML Std", "PY Std"])

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[20, 22, 24, 26, 28, 30]

plt.hist([ml_std_age,py_std_age], bins, rwidth=0.8, color=["r", "y"], label=["ML Std", "PY Std"])

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[20, 22, 24, 26, 28, 30]

plt.hist([ml_std_age,py_std_age], bins, rwidth=0.8, color=["r", "y"], label=["ML Std", "PY Std"])

plt.title("ML Students Age", fontsize=20)
plt.xlabel("Student Age", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
plt.legend()
plt.grid(color="r")
```
