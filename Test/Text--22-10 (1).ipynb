```python
from sklearn.feature_extraction.text import CountVectorizer

stmt_docs=[row['sentences'] for index,row in df1.iterrows() if row['class'] == 'sentence' ]


stmt_docs
```

```python
vec_s=CountVectorizer()
x_s=vec_s.fit_transform(stmt_docs)
# print(x_s)

tdm_s=pd.DataFrame(x_s.toarray(),columns=vec_s.get_feature_names())
tdm_s
```

```python
stmt_prob=[]
for fre in tdm_s.sum():
  stmt_prob.append(fre/len(stmt_docs))
```

```python
stmt_df=pd.DataFrame({'features':tdm_s.columns.tolist(),'frequency':tdm_s.sum(),'probability':stmt_prob})
stmt_df.reset_index(drop=True,inplace=True)
stmt_df
```

```python
from nltk.tokenize import word_tokenize
new_stmt="what is the price of book"
new_word_list=word_tokenize(new_stmt)
print(new_word_list)
```

```python
prob_s_with_ls = []
for word in new_word_list:
    if word in stmt_df['features'].tolist():
        value = list(stmt_df[stmt_df['features']==word]['probability'])
        count=value[0]
    else:
        count = 0
    prob_s_with_ls.append((count + 1 )/(stmt_len + len_total_feature))
```

```python
new_word_list
```

```python
prob_s_with_ls
```

```python
stmt_prob=dict(zip(new_word_list,prob_s_with_ls))
stmt_prob
```

```python
total_stmt_prob=1
for values in stmt_prob.values():
  total_stmt_prob=total_stmt_prob * values

print(total_stmt_prob)  
```

```python
prob_new_stmt= (total_stmt_prob * (stmt_len/len_total_feature))/(ques_len/len_total_feature)
prob_new_stmt 
```

```python
prob_q_with_ls = []
for word in new_word_list:
    if word in ques_df['features'].tolist():
        value = list(ques_df[ques_df['features']==word]['probability'])
        count=value[0]
    else:
        count = 0
    prob_q_with_ls.append((count + 1)/(ques_len + len_total_feature))
```

```python
ques_prob=dict(zip(new_word_list,prob_q_with_ls))
ques_prob 
```

```python
total_ques_prob=1
for values in ques_prob.values():
  total_ques_prob=total_ques_prob * values

print(total_ques_prob)  
```
