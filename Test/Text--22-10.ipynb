```python
from sklearn.feature_extraction.text import CountVectorizer

stmt_docs=[row['sentences'] for index,row in df1.iterrows() if row['class'] == 'sentence' ]


stmt_docs
```

```python
for index,row in df1.iterrows():
  type(row)
  # print(row) 
```

```python
vec_s=CountVectorizer()
x_s=vec_s.fit_transform(stmt_docs)
# print(x_s)

tdm_s=pd.DataFrame(x_s.toarray(),columns=vec_s.get_feature_names())
tdm_s
```

```python
ques_docs=[rows['sentences'] for index ,rows in df1.iterrows() if rows['class']=='question']
ques_docs
```
