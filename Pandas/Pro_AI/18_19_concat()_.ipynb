```python
import pandas as pd
```

```python
sr1= pd.Series([1, 2, 3, 4, 5])
sr1
```

```python
sr2=pd.Series([6, 7, 8, 9, 10])
sr2
```

```python
df1=pd.DataFrame({
    "Id":[1, 2, 3, 4, 5],
    "Name":["Harsh", "Krunal", "Karan", "Himanshu", "Hashmukh"],
})
df1
```

```python
df2=pd.DataFrame({
    "Id":[1, 2, 3, 4, 6],
    "Age":[22, 21, 23, 24, 24]
})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], join="inner")
```

```python
pd.concat([df1, df2], keys=["df1", "df2"])
```

```python
pd.concat([df1, df2], axis=1, keys=["df1", "df2"])
```
